```python
# Lista per memorizzare i viaggi , con alcune linee annidate per raggruppare i viaggi per nazioni
viaggi = {
    "Italia": [
        {"Citt√†": "Roma", "Mare": False, "Prezzo": 700, "Attrazioni": ["Colosseo", "Fontana di Trevi"]},
        {"Citt√†": "Firenze", "Mare": False, "Prezzo": 600, "Attrazioni": ["Ponte Vecchio", "Galleria degli Uffizi"]}
    ],
    "Francia": [
        {"Citt√†": "Parigi", "Mare": False, "Prezzo": 1200, "Attrazioni": ["Torre Eiffel", "Louvre"]},
        {"Citt√†": "Nizza", "Mare": True, "Prezzo": 500, "Attrazioni": ["Promenade des Anglais", "Colline du Ch√¢teau"]}
    ],
    "Spagna": [
        {"Citt√†": "Barcellona", "Mare": True, "Prezzo": 450, "Attrazioni": ["Sagrada Familia", "Parc G√ºell"]}
    ],
    "Grecia": [
        {"Citt√†": "Santorini", "Mare": True, "Prezzo": 350, "Attrazioni": ["Spiagge", "Caldera"]}
    ],
    "Regno Unito": [
        {"Citt√†": "Londra", "Mare": False, "Prezzo": 820, "Attrazioni": ["Big Ben", "Buckingham Palace"]}
    ],
    "Portogallo": [
        {"Citt√†": "Lisbona", "Mare": True, "Prezzo": 300, "Attrazioni": ["Torre di Belem", "Alfama"]}
    ],
    "Germania": [
        {"Citt√†": "Berlino", "Mare": False, "Prezzo": 550, "Attrazioni": ["Muro di Berlino", "Porta di Brandeburgo"]}
    ],
    "Olanda": [
        {"Citt√†": "Amsterdam", "Mare": False, "Prezzo": 740, "Attrazioni": ["Museo Van Gogh", "Canali"]}
    ],
    "Turchia": [
        {"Citt√†": "Istanbul", "Mare": True, "Prezzo": 210, "Attrazioni": ["Moschea Blu", "Santa Sofia"]}
    ],
    "Stati Uniti": [
        {"Citt√†": "New York", "Mare": False, "Prezzo": 1950, "Attrazioni": ["Statua della Libert√†", "Times Square"]}
    ]
}

# Funzione per aggiungere un nuovo viaggio
def aggiungi_viaggio(nazione, citt√†, mare, prezzo, attrazioni):
    nuovo_viaggio = {
        "Citt√†": citt√†,
        "Mare": mare,
        "Prezzo": prezzo,
        "Attrazioni": attrazioni
    }
    
    # Se la nazione esiste gi√†, aggiungi la citt√† alla lista
    if nazione in viaggi:
        viaggi[nazione].append(nuovo_viaggio)
    else:
        # Se la nazione non esiste, creala con la nuova citt√†
        viaggi[nazione] = [nuovo_viaggio]
    
    print(f"Viaggio a {citt√†} ({nazione}) aggiunto con successo!\n")
    print()
    print("üëç Viaggio aggiunto con successo!\n")

# Funzione per visualizzare tutti i viaggi
def visualizza_viaggi():
    print("\nLista dei viaggi disponibili:\n")
    for nazione, citt√†_lista in viaggi.items():  # Itera sulle nazioni e le relative citt√†
        for viaggio in citt√†_lista:  # Itera sulle citt√† nella lista
            print(f"Nazione: {nazione}, Citt√†: {viaggio['Citt√†']}, Mare: {'S√¨' if viaggio['Mare'] else 'No'}, Prezzo: {viaggio['Prezzo']}‚Ç¨, Attrazioni: {', '.join(viaggio['Attrazioni'])}")
    print("\n")

# Funzione per filtrare i viaggi per prezzo
def filtra_viaggi_per_prezzo(prezzo_min, prezzo_max):
    print()  # Empty line before list
    found = False  # Variabile per verificare se sono stati trovati viaggi

    # Scorre ogni nazione e ogni citt√† all'interno delle nazioni
    for nazione, citt√†_list in viaggi.items():
        for citt√† in citt√†_list:
            if citt√†['Prezzo'] >= prezzo_min and citt√†['Prezzo'] <= prezzo_max:
                found = True
                print(f"Nazione: {nazione}, Citt√†: {citt√†['Citt√†']}, Mare: {'S√¨' if citt√†['Mare'] else 'No'}, Prezzo: {citt√†['Prezzo']}‚Ç¨, Attrazioni: {', '.join(citt√†['Attrazioni'])}")

    if not found:
        print("Nessun viaggio trovato con il prezzo compreso tra " + str(prezzo_min) + " euro e " + str(prezzo_max) + " euro.")
    print()  # Empty line after list

# Funzione per trovare i viaggi al mare e non
def filtra_viaggi_per_mare(mare):
    print()
    print()  # Empty line before list
    found = False  # Variabile per verificare se sono stati trovati viaggi
    for nazione, citta_list in viaggi.items():  # Iteriamo sulle nazioni
        for citta in citta_list:  # Iteriamo sulle citt√† per ogni nazione
            if citta['Mare'] == mare:
                found = True
                print(f"Nazione: {nazione}, Citt√†: {citta['Citt√†']}, Mare: {'S√¨' if citta['Mare'] else 'No'}, Prezzo: {citta['Prezzo']}‚Ç¨, Attrazioni: {', '.join(citta['Attrazioni'])}")
    if not found:
        print("Nessun viaggio trovato per questa opzione.")
    print()
    print()  # Empty line after list 

# Funzione per trovare il viaggio al mare pi√π economico
def filtra_viaggi_mare_piu_economico():
    # Itera attraverso tutte le nazioni e i viaggi all'interno di ogni nazione
    viaggi_mare = [
        {"Nazione": nazione, "Citt√†": viaggio["Citt√†"], "Prezzo": viaggio["Prezzo"]}
        for nazione, viaggi in viaggi.items()
        for viaggio in viaggi
        if viaggio["Mare"]
    ]
    
    if viaggi_mare:
        
        # Trova il viaggio al mare pi√π economico
        viaggio_economico = min(viaggi_mare, key=lambda v: v["Prezzo"])
        print(f"Viaggio pi√π economico al mare: {viaggio_economico['Nazione']}, {viaggio_economico['Citt√†']}, Prezzo: {viaggio_economico['Prezzo']}‚Ç¨")
    else:
        print("Nessun viaggio disponibile al mare.")

#Funzione per ordinare i viaggi per prezzo dal pi√π economico al pi√π costoso
def ordina_viaggi_per_prezzo():
    # Creiamo una lista di tutti i viaggi
    viaggi_ordinati = []
    
    # Iteriamo su ogni nazione
    for nazione, citt√†_list in viaggi.items():
        for citt√† in citt√†_list:
            viaggio = {"Nazione": nazione, "Citt√†": citt√†["Citt√†"], "Prezzo": citt√†["Prezzo"]}
            viaggi_ordinati.append(viaggio)

    # Ordiniamo la lista di viaggi per prezzo crescente
    viaggi_ordinati = sorted(viaggi_ordinati, key=lambda v: v['Prezzo'])

    # Stampiamo i viaggi ordinati
    print("Viaggi ordinati per prezzo crescente:")
    for viaggio in viaggi_ordinati:
        print(f"Nazione: {viaggio['Nazione']}, Citt√†: {viaggio['Citt√†']}, Prezzo: {viaggio['Prezzo']}‚Ç¨")

# Funzione per filtrare i viaggi al mare, con pi√π criteri come Nazione e fascia di prezzo 
def filtra_viaggi_mare_nazione_prezzo(nazione, prezzo_min, prezzo_max):
    print(f"Viaggi al mare in {nazione} con prezzo tra {prezzo_min}‚Ç¨ e {prezzo_max}‚Ç¨:")
    found = False
    if nazione in viaggi:  # Verifica che la nazione esista nel dizionario
        for viaggio in viaggi[nazione]:
            if viaggio['Mare'] and prezzo_min <= viaggio['Prezzo'] <= prezzo_max:
                found = True
                print(f"Citt√†: {viaggio['Citt√†']}, Prezzo: {viaggio['Prezzo']}‚Ç¨, Attrazioni: {', '.join(viaggio['Attrazioni'])}")
    if not found:
        print("Nessun viaggio trovato con i criteri specificati.")    

# Funzione per contare quanti viaggi ci sono per ogni Nazione
def conta_viaggi(viaggi):
    # Creiamo un dizionario per memorizzare i contatori
    conteggio_viaggi = {}

    # Iteriamo attraverso il dizionario delle nazioni
    for nazione, destinazioni in viaggi.items():
        # Conta quante citt√† ci sono per ogni nazione
        conteggio_viaggi[nazione] = len(destinazioni)

    return conteggio_viaggi   

        
# Esecuzione del programma
def main():
    while True:
        print()
        print("1. Aggiungi viaggio")
        print("2. Visualizza tutti i viaggi che abbiamo")
        print("3. Filtra viaggi per prezzo")
        print("4. Filtra per destinazione di mare o meno")
        print("5. Visualizza il viaggio pi√π economico al mare")
        print("6. Visualizza tutti i viaggi dal pi√π economico al pi√π costoso")
        print("7. Filtra le localit√† di mare in base alla Nazione e alla fascia di prezzo")
        print("8. Tieni il conto di quanti viaggi abbiamo per Paese")
        print("9. Esci")
        print()
        scelta = input("Scegli un'opzione: ")

        if scelta == "1":
            nazione = input("Inserisci la nazione: ")
            citt√† = input("Inserisci la citt√†: ")
            mare  = input("√à una destinazione di mare? (s√¨/no): ").strip().lower()
            if mare in ["s√¨", "si"]:
                    mare = True
            elif mare == "no":
                    mare = False
            else:
                print("Risposta non valida. Inserisci 's√¨' o 'no'.")
            prezzo = float(input("Inserisci il prezzo: "))
            attrazioni = input("Inserisci le attrazioni (separate da virgola): ").split(",")
            aggiungi_viaggio(nazione, citt√†, mare, prezzo, [a.strip() for a in attrazioni])

        elif scelta == "2":
            visualizza_viaggi()

        elif scelta == "3":
            prezzo_min = float(input("Inserisci il prezzo minimo: "))
            prezzo_max = float(input("Inserisci il prezzo massimo: "))
            filtra_viaggi_per_prezzo(prezzo_min, prezzo_max)

        elif scelta == "4":
            risposta = input("Vuoi una destinazione di mare? (s√¨/no): ").strip().lower()
            if risposta == "s√¨":
                filtra_viaggi_per_mare(True)
            elif risposta == "no":
                filtra_viaggi_per_mare(False)
            else:
                print("Risposta non valida. Riprova.")

        elif scelta == "5":  
            filtra_viaggi_mare_piu_economico()

        elif scelta == "6":
            ordina_viaggi_per_prezzo()
        
        elif scelta == "7":
            nazione = input("Inserisci la nazione: ")
            prezzo_min = float(input("Inserisci il prezzo minimo: "))
            prezzo_max = float(input("Inserisci il prezzo massimo: "))
            filtra_viaggi_mare_nazione_prezzo(nazione,prezzo_min, prezzo_max)
        
        elif scelta == "8":
            conta_viaggi(viaggi)
            conteggio = conta_viaggi(viaggi)
            print(conteggio) 
        
        elif scelta == "9":
            print ("üëã Ciao alla prossima e Buona Giornata!")
            break
        
if __name__ == "__main__":
    main()
```
